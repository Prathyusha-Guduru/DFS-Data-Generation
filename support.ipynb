```python
import numpy as np
import pandas as pd
```

```python
excelFile = pd.ExcelFile('DFS_project.xlsx')
sheetNames = excelFile.sheet_names
sheetNames
```

```python
ex = excelFile.parse(sheetNames[1])
ex.head()
```

```python
#Renaming column heads appropriatelys

ex.columns = [*ex.columns[:-1], 'ts']
ex = ex.rename(columns={ex.columns[0]:'rack_no', ex.columns[1]:'d_no',ex.columns[2]:'file_no'})

for column_head in range(3,len(ex.columns)-1):
	ex = ex.rename(columns = {ex.columns[column_head] : f"block_{column_head-3}"})

ex.head()
```

```python
# Calculating support for each file

file_count = np.array(ex['file_no'].value_counts())
log_length = len(ex)
support = np.true_divide(file_count,log_length)
print(support)

```
